```python
def plot_metrics(axs, metrics,color1,color2):
    max_len = len(metrics)
    for i in range(max_len):
        axi = axs[int(i/5),i%5]
        axi.set_title('Lambda : '+str(metrics[i]["m_lambda"]))
        axi.set_xlabel('N')
        x = metrics[i]["ns"]
        y = metrics[i]["f1"]
        y1 = metrics[i]["f1"]
        axi.scatter(x,y,c=color1)
        axi.scatter(x,y1,c=color2)
        
```

```python
fig, axs = plt.subplots(2, 5,figsize=(25, 10))
# axes = plt.gca()
# plt.ylim(0,1)
plot_metrics(axs , MetricsUtil.get_metrics_data(case_name,purpose), 'blue','red')
plt.show()

fig, axs = plt.subplots(2, 5,figsize=(25, 10))
plot_metrics(axs , MetricsUtil.get_metrics_data(case_name,"test"),'blue','red')
plt.show()
```
