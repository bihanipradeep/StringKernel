```python
def plot_metrics(axs, case_name, purpose,color1,color2):
    metrics = MetricsUtil.get_metrics_data(case_name, purpose)
    max_len = len(metrics)
    for i in range(max_len):
        axs.set_title(purpose)
        axs.set_xlabel('N')
        x = metrics[i]["ns"]
        y = metrics[i]["f1_1"]
        y1 = metrics[i]["f1_2"]
        axs.scatter(x,y,c=color1)
        axs.scatter(x,y1,c=color2)
        for j in range(len(metrics[i]["ns"])):
            axs.annotate(str(metrics[i]["m_lambda"]), (x[j], y[j]))
            axs.annotate(str(metrics[i]["m_lambda"]), (x[j], y1[j]))
```

```python
fig, axs = plt.subplots(1,2,figsize=(10, 5))
plot_metrics(axs[0] ,case_name, purpose, 'blue','red')
plot_metrics(axs[1] ,case_name, "test", 'blue','red')
plt.show()

# fig, axs = plt.subplots(2, 5,figsize=(25, 10))
# plot_metrics(axs , MetricsUtil.get_metrics_data(case_name,"test"),'blue','red')
# plt.show()
```
